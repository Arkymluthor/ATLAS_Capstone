```python
df_agg = pd.read_csv('healthkey.csv',index_col=0)
df_agg.head()
```

```python
def appender(dataframe):
    agg_ = df_agg.merge(dataframe[['Unit','People_Vaccinated']],on='Unit')
    
    return agg_
```

```python
agg_ = df_agg.merge(df1[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df1'})
agg_ = agg_.merge(df2[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df2'})
agg_ = agg_.merge(df3[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df3'})
agg_ = agg_.merge(df4[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df4'})
agg_ = agg_.merge(df5[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df5'})
agg_ = agg_.merge(df6[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df6'})
agg_ = agg_.merge(df7[['Unit','People_Vaccinated']],on='Unit').rename(columns={'People_Vaccinated': 'People_vac_df7'})
```

```python
agg_ 
```

```python
agg_=agg_.replace('No Data','0')
```

```python
def split_convert(x):
    if len(x) <= 3:
        return int(x)

    else:
        return int(x.replace(',',''))
```

```python
agg_['People_vac_df1'] = agg_['People_vac_df1'].apply(split_convert)
agg_['People_vac_df2'] = agg_['People_vac_df2'].apply(split_convert)
agg_['People_vac_df3'] = agg_['People_vac_df3'].apply(split_convert)
agg_['People_vac_df4'] = agg_['People_vac_df4'].apply(split_convert)
agg_['People_vac_df5'] = agg_['People_vac_df5'].apply(split_convert)
agg_['People_vac_df6'] = agg_['People_vac_df6'].apply(split_convert)
agg_['People_vac_df7'] = agg_['People_vac_df7'].apply(split_convert)

```
