```python
url = 'http://www.city-data.com/canada/Ontario-Health-regions.html'
```

```python
driver.get(url)
```

```python
table_container = driver.find_elements_by_class_name('r1')
 # Sleep
time.sleep(4)

```

```python
Health_Unit={}
for container in table_container:
    #print(container.find_element_by_tag_name('td').text)
    row = container.find_elements_by_tag_name('td')
    Health_Unit[row[0].text] = row[1].text

    
```

```python
links=[]
for container in table_container:
    #print(container.find_element_by_tag_name('td').text)
    row = container.find_elements_by_tag_name('a')
    for item in row:
        links.append(item.get_attribute('href'))
```

```python
links[:5]
```

```python
Health_Reg=[] #Health region
Region=[] #sub region
Seg =[] #city,town, or village
pop =[] #population

for link in links:
    driver1 = webdriver.Chrome(chrome_path)
    driver1.get(link)
    row_content = driver1.find_elements_by_xpath("//table[@id='cityTAB']/tbody/tr")
    Health_Reg.append(len(row_content))
    for content in row_content:
        items = content.find_elements_by_tag_name('td')
        Region.append(items[-3].text)
        Seg.append(items[-2].text)
        pop.append(items[-1].text)
    driver1.close()
    
```
